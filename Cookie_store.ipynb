```python
#Import necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
```

```python
#incase of unnecessary errors
warnings.filterwarnings("ignore", category=pd.core.common.SettingWithCopyWarning)

#Import data
Cookie_2017=pd.DataFrame(pd.read_csv('C:/Program Files/PostgreSQL/15/data/data_copy/2017_Order_Data.csv'))
Cookie_2018=pd.DataFrame(pd.read_csv('C:/Program Files/PostgreSQL/15/data/data_copy/2018_Order_Data.csv'))
Cookie_2019=pd.DataFrame(pd.read_csv('C:/Program Files/PostgreSQL/15/data/data_copy/2019_Order_Data.csv'))
Cookie_2020=pd.DataFrame(pd.read_csv('C:/Program Files/PostgreSQL/15/data/data_copy/2020_Order_Data.csv'))

# Get the first 5 values in Cookie_2017
Cookie_2017.head()
```

```python
# Get the last 5 values in Cookie_2018
Cookie_2018.tail()
```

```python
# Get 5 random values in Cookie_2019
Cookie_2019.sample(5)
```

```python
# Get 5 random values in Cookie_2020
Cookie_2020.sample(5)
```

```python
# Create temp table
both_table = pd.concat([Cookie_2017,Cookie_2018,Cookie_2019,Cookie_2020])
both_table.describe(include='all')
```

```python
# Count null 
both_table.isnull().sum()
```

```python
# Drop duplicates
both_table.drop_duplicates(inplace=True)
#Count Null
both_table.isnull().sum()
```

```python
# Drop Null in any column
both_table.dropna(how='any', inplace=True)

# Count if there are till null values
both_table.isnull().sum()
```

```python
# Describe all the columns both integer and non integer
both_table.describe(include='all')
```

```python
# Create new columns from splitting Customer_ID
both_table['Customer_Name'] = both_table['Customer_ID'].str.split(' - ').str[1]

both_table['Customer_ID'] = both_table['Customer_ID'].str.split(' - ').str[0]


both_table
```

```python
# Change data types of columns
both_table[['Cookies_Shipped','Revenue','Cost']] = both_table[['Cookies_Shipped','Revenue','Cost']].astype(int)
both_table[['Order_ID', 'Customer_Name']] = both_table[['Order_ID','Customer_Name']].astype('object')
both_table
```

```python
# Calculate the profit
both_table['Profit'] = both_table['Revenue'] - both_table['Cost']
both_table
```

```python
# Convert columns to date and check with info() to be sure
both_table['Order_Date'] = pd.to_datetime(both_table['Order_Date'])
both_table['Ship_Date'] = pd.to_datetime(both_table['Ship_Date'])

both_table.info()
```

```python
# Extract year from date column
both_table['Order_Year'] = both_table['Order_Date'].dt.year
both_table['Ship_Year'] = both_table['Ship_Date'].dt.year

# Change the data type
both_table[['Order_Year', 'Ship_Year']] = both_table[['Order_Year','Ship_Year']].astype('object')

# Find the data difference
both_table['Diff'] = (both_table['Ship_Date'] - both_table['Order_Date']).dt.days

# Get information about the table
both_table.info()
```

```python
#Drop Order_Status column
both_table=both_table.drop(columns='Order_Status')
both_table
```

```python
# Restructure table
both_table = both_table[['Order_ID', 'Customer_ID', 'Customer_Name', 'Cookies_Shipped','Revenue','Cost', 
                         'Profit','Order_Date','Ship_Date','Diff','Order_Year','Ship_Year']]

both_table
```

```python
value = both_table.groupby(['Customer_Name','Order_Year'])[['Profit']].agg(sum).reset_index()
sns.catplot(x='Customer_Name', y='Profit',data=value, hue='Order_Year',kind='bar',palette="Blues").set_xticklabels(rotation=90)
sns.set_style("whitegrid")

# plt.figure(figsize=(22, 6))
plt.title("Stacked Bar Chart",fontsize=15, fontweight='bold')
plt.xlabel("Customer Names", fontsize=12)
plt.ylabel("Total Profit", fontsize=12)

plt.show()
both_table.groupby(['Customer_Name','Order_Year'])[['Profit']].agg(sum)

# The analysis shows that the most cookies are bought by 'Cascade Grovers'.
# But in 2020 the profit of 'Cascade Grovers' dropped to 256466. 
# While the least cookies purchased from 2017-2020 are by 'Acme Grocery Stores'
```

```python
value=both_table.groupby('Order_Year')[['Profit','Cookies_Shipped']].agg('sum').sort_values(by = 'Order_Year',ascending=False).reset_index()

sns.set_style("whitegrid")
g=sns.catplot(x='Order_Year', y='Cookies_Shipped',data=value,kind='bar', palette="Blues")
h=sns.catplot(x='Order_Year', y='Profit',data=value,kind='bar',palette="Blues")

g.set_axis_labels("Order Year", "Cookies Shipped", fontsize = 15)
h.set_axis_labels("Order Year", "Profit", fontsize = 15)

plt.figure(figsize=(20, 6))

g.fig.suptitle('Cookies Shipped by Order Year',y=1.03, fontsize=15, fontweight='bold')
h.fig.suptitle('Profit by Order Year',y= 1.03 ,fontsize=15, fontweight='bold')

h.ax.get_yaxis().get_major_formatter().set_scientific(False)


plt.show()

both_table.groupby('Order_Year')[['Cookies_Shipped','Profit']].agg('sum').sort_values(by = 'Order_Year',ascending=False)

# The most cookies and profit year is 2018, while the demand for cookies dropped in 2020 and so did the profit
```

```python
both_table.groupby(['Customer_ID', 'Customer_Name']) [['Cookies_Shipped', 'Profit']].agg('sum')

# customer with id 325698 (Cascade Grovers) have brought the most cookies and
# have generated the most profit for kelvin's cookies 
```

```python
value= both_table.groupby(['Customer_Name', 'Order_Year']) [['Cost', 'Profit']].agg('sum').reset_index()
sns.set_style("whitegrid")
g=sns.catplot(x='Customer_Name', y='Cost',data=value,kind='bar', hue= 'Order_Year', palette="Blues").set_xticklabels(rotation=90)
h=sns.catplot(x='Customer_Name', y='Profit',data=value,kind='bar', hue='Order_Year', palette="Blues").set_xticklabels(rotation=90)

g.set_axis_labels("Customer_Name", "Cost", fontsize = 15)
h.set_axis_labels("Customer_Name", "Profit", fontsize = 15)

plt.figure(figsize=(20, 6))

g.fig.suptitle('Cost by Customer_Name',y=1.03, fontsize=15, fontweight='bold')
h.fig.suptitle('Profit by Customer_Name',y= 1.03 ,fontsize=15, fontweight='bold')

h.ax.get_yaxis().get_major_formatter().set_scientific(False)


plt.show()


both_table.groupby(['Customer_Name', 'Order_Year']) [['Cost', 'Profit']].agg('sum')

# The man reason for profit drop in 2020 is because of reduction in cost. The eagerness for this product shows how valuable 
# it is whether sold at a high rate or low rate.
```
